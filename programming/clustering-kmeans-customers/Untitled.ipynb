```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import sklearn
```

```python
dataset = pd.read_csv('customers.csv')
X = dataset.iloc[:, [3, 4]].values

fig = plt.figure()
ax = plt.subplot()
ax.plot( X[:,0]
       , X[:,1]
       , "."
       )
ax.set_xlabel("Customer Income $K")
ax.set_ylabel("Customer Spending Score")
plt.savefig("customers.png")
```

```python
from sklearn.cluster import KMeans
wcsds = []
for i in range(1, 11): 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X) 
    wcsds.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('Within-Cluster Sum of Distances-Squared') 
plt.tight_layout()
plt.show()
```
