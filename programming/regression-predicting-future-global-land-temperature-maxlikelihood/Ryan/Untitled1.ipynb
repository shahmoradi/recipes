```python
import matplotlib.pyplot as plt
import numpy as np
x_values = np.double([-1, 0, 1])
for slope in np.linspace(-10,10,101):
    y_values = slope * np.double(x_values)
    #plt.plot(x_values, y_values, "blue", linewidth=5, alpha=0.3)
    plt.plot(x_values, y_values, "red", linewidth=1, alpha=0.1)
```
